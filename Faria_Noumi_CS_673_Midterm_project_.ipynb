```python
# Check for outliers using IQR
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
IQR = q3 - q1

not_outliers = df[~((df < (q1 - 1.5 * IQR)) | (df > (q3 + 1.5 * IQR)))]

# Outliers filled with zeros.
New_df = not_outliers.fillna(value=np.nan)


# Display the first 10 rows of the new DataFrame
New_df.head(10)


```

```python
import pandas as pd
import sqlite3
import os
from sqlalchemy import create_engine

# Create an engine object for SQLite
engine = create_engine('sqlite:///my_healthcare_database.db')

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/CS_673_Dataset/Healthcare_Diabetes.csv')

# Insert the DataFrame into the database using the engine
# if_exists='replace' will replace the table if it exists
df.to_sql('users', engine, if_exists='replace', index=False)

# Now, you can read data from the database using the same engine
query = "SELECT * FROM users"
df_from_db = pd.read_sql(query, engine)

print(df_from_db.head())
```
